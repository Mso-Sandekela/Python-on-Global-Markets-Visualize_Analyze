```python
#_____________________________________________________________
#Importing libraries...
import yfinance as yf
import mplfinance as mpf
import pandas as pd

ticker = input("Enter the Stock name: ").upper().strip() #get stock name from user

#Download stock data,I specified the range I need  04 Aug--3 Oct
df = yf.download(ticker, start='2025-08-04', end='2025-10-03') 


if df.empty: #If it happens that the data is not found.
    print(f"No data found for {ticker} in the given date range.")
else:
    #If df has MultiIndex columns, flatten them
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [col[0] for col in df.columns]

    #Keep only OHLCV
    df = df[["Open", "High", "Low", "Close", "Volume"]]

    #Ensure numeric.
    df = df.apply(pd.to_numeric, errors="coerce").dropna()

    if df.empty: #if df empty
        print(f"After cleaning, no valid OHLC data available for {ticker}.")
    else:
        # Add a key/label inside the chart using mpf.make_addplot
        ap = [mpf.make_addplot([df['Close'].iloc[-1]]*len(df), type='line', color='blue', width=0.5)]
        
        # Plot candlestick chart with title as key
        mpf.plot(df, type="candle", style="charles",
                 title=f"{ticker} Candlestick Chart",
                 ylabel="Price",
                 addplot=ap)
#_____________________________________________________________
# Visit My Git Repository --> https://github.com/Mso-Sandekela
#               FOR MORE AWSWOME CODES
#_____________________________________________________________
#=============================================================
```

```python
#_____________________________________________________________
# Importing libraries...
import yfinance as yf
import mplfinance as mpf
import pandas as pd

# Ask user for 4 tickers
tickers = input("Enter 4 Stock/Commodity tickers separated by spaces: ").upper().strip().split()

# Ensure exactly 4 tickers
if len(tickers) != 4:
    print("Please enter exactly 4 tickers separated by spaces.")
else:
    for ticker in tickers:
        print(f"\nFetching data for: {ticker}")
        df = yf.download(ticker, start='2025-08-04', end='2025-10-03')
        
        if df.empty:
            print(f"No data found for {ticker} in the given date range.")
            continue
        
        # Flatten MultiIndex columns if needed
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = [col[0] for col in df.columns]
        
        # Keep only OHLCV
        df = df[["Open", "High", "Low", "Close", "Volume"]]
        
        # Convert to numeric and drop invalid rows
        df = df.apply(pd.to_numeric, errors="coerce").dropna()
        
        if df.empty:
            print(f"After cleaning, no valid OHLC data available for {ticker}.")
            continue
        
        # Add a key/label inside the chart using `mpf.make_addplot`
        ap = [mpf.make_addplot([df['Close'].iloc[-1]]*len(df), type='line', color='blue', width=0.5)]
        
        # Plot candlestick chart with title as key
        mpf.plot(df, type="candle", style="charles",
                 title=f"{ticker} Candlestick Chart",
                 ylabel="Price",
                 addplot=ap)
#_____________________________________________________________
# Visit My Git Repository --> https://github.com/Mso-Sandekela
#               FOR MORE AWESOME CODES
#_____________________________________________________________
#=============================================================
```
